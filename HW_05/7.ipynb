```python
import pandas as pd
import numpy as np

df = pd.read_csv("iris_dataset.csv")
features = df.columns[df.columns != "target"]
df_feature=df[features]
df_feature
```

```python
pd.set_option('display.expand_frame_repr', False)

group = df.groupby("target")
mean = group.agg("mean")
std = group.agg("std")
print("mean")
print(mean)
print("std")
print(std)
print("range")
print(group.agg(lambda x: np.max(x)-np.min(x)))
print("max")
print(group.agg("max"))
print("min")
print(group.agg("min"))
```
