```python
import pandas as pd

df = pd.read_csv("GlobalLandTemperaturesByCountry.csv")
print(df.dtypes)
print(df.shape)
print(df.describe())
```

```python
df.isna().sum()
```

```python
droped_df = df.dropna(inplace=False)
droped_df
```

```python
moving_avg = df.select_dtypes("number").rolling(60,min_periods=1,center=True).mean()
filled_df = df.fillna(moving_avg ,inplace=False)
filled_df
```

. افزودن ستون Decade :
تبدیل سالها به دههها. به عنوان مثال، سال 1980 به “ 1980s

```python
filled_df.dt = pd.to_datetime(filled_df.dt)
filled_df["Decade"] = (filled_df.dt.dt.year // 10 * 10).astype(str) + "s"
filled_df
```

نگهداری فقط کشورهایی که حداقل 100 سال داده دارند.

```python
year_count = filled_df.groupby("Country").dt.agg(lambda x: max(x).year-min(x).year)
# year_count = pd.to_datetime(year_count.max().dt) - pd.to_datetime(year_count.min().dt)
print(year_count[year_count<=100])
filtered_countries = filled_df[filled_df["Country"].isin(year_count[year_count>100].index.tolist())]
filtered_countries
# print(filtered_countries[filtered_countries.Country == "Antarctica"])
```

محاسبه میانگین دما برای هر دهه و ساخت ماتریس decade × temperature anomaly .

```python
grouped = filtered_countries.groupby(["Country","Decade"]).mean(numeric_only=True)
grouped
```

استفاده از np.polyfit )( برای محاسبه خط روند دما.

```python
import numpy as np
import matplotlib.pyplot as plt

decades = grouped.loc["Afghanistan"].index
temps   = grouped.loc["Afghanistan"]["AverageTemperature"]

x_numeric = np.array([int(dec[:-1]) + 5 for dec in decades])
polyfit = np.polyfit(x_numeric, temps, 1)
polyfit
```

محاسبه نرخ گرمایش بر حسب ° C بر دهه.
