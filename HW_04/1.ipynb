```python
import pandas as pd

df = pd.read_csv(
    "iris.data",
    header=None,
    names=["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
)
print(df.shape)
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
df_corr=df.corr(numeric_only=True)
df_corr
```

```python
df["petal_ratio"]=df.petal_length/df.petal_width
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.heatmap(df_corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
sns.pairplot(df ,hue="species")
plt.show()
```

```python
sns.boxplot(df, x="species", y= "petal_ratio")
plt.show()
```

```python
sns.scatterplot(df, x="petal_length", y="petal_ratio", hue="species")
plt.grid(True)
plt.show()
```
